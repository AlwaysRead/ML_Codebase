```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt

california_housing=fetch_california_housing(as_frame=True)

min_ave_bedrms=california_housing.data['AveBedrms'].min()
max_ave_bedrms=california_housing.data['AveBedrms'].max()
mean_ave_bedrms=california_housing.data['AveBedrms'].mean()
median_ave_bedrms=california_housing.data['AveBedrms'].median()

print(f"Features: Bedrooms")
print(f"Min Value: {min_ave_bedrms}")
print(f"Max Value: {max_ave_bedrms}")
print(f"Mean Value: {mean_ave_bedrms}")
print(f"Median Value: {median_ave_bedrms}")

n_features=len(california_housing.data.columns)

fig, axes=plt.subplots(n_features,1,figsize=(8,2*n_features))

if n_features ==1:
    axes=[axes]
for i, feature in enumerate(california_housing.data.columns):
    axes[i].hist(california_housing.data[feature],color='skyblue',edgecolor='black')
    axes[i].set_title(f'Features:{feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
